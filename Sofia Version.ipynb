```python
# Dependencies
import csv
import pandas as pd
import os
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress 
import seaborn as sns 
from scipy import stats
import matplotlib.dates as mdates
import scipy.stats as st
from pprint import pprint
from datetime import datetime
```

```python
# Import the covid-19 dataset 
csv_path = ('Resources/owid-covid-data.csv')
covid_data_df = pd.read_csv(csv_path)
covid_data_df
```

```python
#Total Cases, Deaths and Mortality by Continent
cases_by_continent = covid_data_df.groupby(["continent"]).agg({'new_cases':['sum'],'new_deaths':['sum']})

cases_by_continent["Mortality_rate"]= cases_by_continent['new_deaths']/ cases_by_continent['new_cases']

cases_by_continent.columns=['Total Cases','Total Deaths', 'Mortality Rate']
cases_by_continent
```

```python
#Top 10 Countries with highest mortality rate
cases_by_country = covid_data_df.groupby(["location"]).agg({'new_cases':['sum'],'new_deaths':['sum']})

cases_by_country["Mortality_rate"]= cases_by_country['new_deaths']/ cases_by_country['new_cases']

cases_by_country.columns=['Total Cases','Total Deaths', 'Mortality Rate']
top=cases_by_country.sort_values('Mortality Rate',ascending=False)
top.head(10)
```

```python
#Average life expectacy for each continent
life_expectancy_by_continent = covid_data_df.groupby(["continent"])['life_expectancy'].mean()

#Average GDP for each continent
avg_gdp_continent=covid_data_df.groupby(['continent'])['gdp_per_capita'].mean()

#Mean Population sensity per continent
pop_density=covid_data_df.groupby(['continent'])['population_density'].mean()

Summary_df=pd.DataFrame({ 'Life Expectancy':life_expectancy_by_continent, 'GDP':avg_gdp_continent, 'Population Density':pop_density})
Summary_df
```

```python
#Merged dataframe

summarized_df=pd.merge(cases_by_continent, Summary_df, how="left", on="continent")
summarized_df
```

```python
x_values = summarized_df['Life Expectancy']
y_values = summarized_df['Mortality Rate']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
#plt.annotate(line_eq,(2,2),fontsize=10,color="black")
plt.xlabel('Life Expectancy')
plt.ylabel('Mortality Rate')
plt.show()
```

```python
x_values = summarized_df['GDP']
y_values = summarized_df['Mortality Rate']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
#plt.annotate(line_eq,(2,2),fontsize=10,color="black")
plt.xlabel('GDP')
plt.ylabel('Mortality Rate')
plt.show()
```

```python
x_values = summarized_df['Population Density']
y_values = summarized_df['Total Cases']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
#plt.annotate(line_eq,(2,2),fontsize=10,color="black")
plt.xlabel('Population Density')
plt.ylabel('Total Cases')
plt.show()
```

```python
#Conclusions:
#Continents with higher life expectancy, experienced higher mortality rate. Which means the eldery population are most likely to die from COVID
#Countries/Continents with higher GDP are more likely to be affected by covid compared to poor countries with lower GDP
#Population density is not a key factor for the high spread of COVID
```
